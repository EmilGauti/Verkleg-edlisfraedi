```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 

from matplotlib.ticker import NullFormatter
plt.ion()



# fastar til að reikna n
q = 1.602*10**(-19)
k_B = 1.38*10**(-23)
T = 298

#1N4148
rev_1N = pd.read_csv('1N4148_rev.txt',sep='\t')
x_1N = rev_1N['Voltage (V)']
y_1N = rev_1N['Current (A)']


#BZX-55C13
rev_BZX = pd.read_csv('BZX-55C13_rev.txt',sep='\t')
x_BZX = rev_BZX['Voltage (V)']
y_BZX = rev_BZX['Current (A)']


#BYS26-45
rev_BYS = pd.read_csv('BYS26-45_rev.txt',sep='\t')
x_BYS = rev_BYS['Voltage (V)']
y_BYS = rev_BYS['Current (A)']


#Blár tvistur
rev_blue = pd.read_csv('blue_rev.txt',sep='\t')
x_BLUE = rev_blue['Voltage (V)']
y_BLUE = rev_blue['Current (A)']

# 1N4148
plt.subplot(221)
plt.scatter(x_1N, y_1N)
plt.yscale('symlog')
plt.title('1N4148')
plt.grid(True, which="both")
plt.xlim(0.9*min(x_1N),1.1*max(x_1N))
plt.ylim(-2e-9, -2e-8)
plt.xlabel('Spenna (V)')
plt.ylabel('Straumur (A)')
plt.gca().invert_yaxis()
plt.locator_params(axis='y', numticks=100)



# BZX-55C13
plt.subplot(222)
plt.scatter(x_BZX,y_BZX)
plt.yscale('symlog')
plt.title('BZX-55C13')
plt.grid(True)
plt.xlim([-15,0])
plt.ylim(-8.0E-11,-9.5E-3)
plt.xlabel('Spenna (V)')
plt.ylabel('Straumur (A)')
plt.gca().invert_yaxis()
plt.locator_params(axis='y', numticks=10)

# BYS26-45
plt.subplot(223)
plt.scatter(x_BYS,y_BYS)
plt.yscale('symlog')
plt.title('BYS26-45')
plt.grid(True)
plt.xlim([-64,0])
plt.ylim(-3.0E-7,-5.0E-5)
plt.xlabel('Spenna (V)')
plt.ylabel('Straumur (A)')
plt.gca().invert_yaxis()
plt.locator_params(axis='y', numticks=8)

# Blár tvistur
plt.subplot(224)
plt.scatter(x_BLUE, y_BLUE)
plt.yscale('symlog')
plt.title('Blár tvistur')

plt.xlim([-43,0])
plt.ylim(-9.0E-13,-1.0E-7)
#plt.yticks(list(plt.yticks()[0]) + [-10.0E-12])
plt.xlabel('Spenna (V)')
plt.ylabel('Straumur (A)')
plt.gca().invert_yaxis()
plt.locator_params(axis='y', numticks=6)
# Format the minor tick labels of the y-axis into empty strings with
# `NullFormatter`, to avoid cumbering the axis with too many labels.
#plt.gca().yaxis.set_minor_formatter(NullFormatter())
# Adjust the subplot layout, because the logit one may take more space
# than usual, due to y-tick labels like "1 - 10^{-3}"
plt.subplots_adjust(top=2, bottom=0, left=0, right=2, hspace=0.25,
                    wspace=0.35)


#plt.show()
plt.savefig('MAIN_A_rev.png', dpi=300, bbox_inches='tight')
```
