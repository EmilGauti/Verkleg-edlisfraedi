```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt


#
def numfmt(x, pos): # your custom formatter function: divide by 100.0
    s = round(1000000*x,3)
    #s = '{}'.format(x * 1000000.0)
    return s

import matplotlib.ticker as tkr     # has classes for tick-locating and -formatting
xfmt = tkr.FuncFormatter(numfmt)    # create your custom formatter function


#fastar
q = 1.602E-19
K_s = 11.8
eps_0 = 8.85*10**(-12)
d = 0.003
A = (d**2)*np.pi/4



# FYRIR BKUN

lidur1 = pd.read_csv('lidur1',sep='\t', skiprows=5)
V_1 = lidur1[list(lidur1)[0]]
C_1 = lidur1[list(lidur1)[2]]

dCdV_1 = np.divide(C_1**(-2),V_1)
dCdV_1 = dCdV_1[1:]
W_1 = K_s*eps_0*A*C_1**(-1)
W_1 = W_1[1:] 
N_A_1 = 2/(q*K_s*eps_0*A**(2)*np.abs(dCdV_1))

plt.subplot(311)
plt.scatter(W_1, N_A_1,s = 3)
plt.yscale('log')
plt.title('Fyrir bkun', fontsize = 20)
plt.xlabel('Dpt ' + r'$[\mu m$]', fontsize = 15)
plt.ylabel('btarttleiki, ' + r'$N_{A} \ [m^{-3}]$', fontsize = 15)
matplotlib.pyplot.grid(True, which="both")
plt.xlim(0.9*min(W_1),1.1*max(W_1))
plt.ylim(1e21, 6e21)
plt.locator_params(axis='y', numticks=8)
plt.gca().xaxis.set_major_formatter(xfmt)

# EFTIR 393 K BKUN
        
lidur2 = pd.read_csv('lidur2',sep='\t', skiprows=5)
V_2 = lidur2[list(lidur2)[0]]
C_2 = lidur2[list(lidur2)[2]]

dCdV_2 = np.divide(C_2**(-2),V_2)
dCdV_2 = dCdV_2[1:]
W_2 = K_s*eps_0*A*C_2**(-1)
W_2 = W_2[1:] 
N_A_2 = 2/(q*K_s*eps_0*A**(2)*np.abs(dCdV_2))

plt.subplot(312)
plt.scatter(W_2, N_A_2,s = 3)
plt.yscale('log')
plt.title('Eftir 393 K bkun', fontsize = 20)
plt.xlabel('Dpt ' + r'$[\mu m$]', fontsize = 15)
plt.ylabel('btarttleiki, ' + r'$N_{A} \ [m^{-3}]$', fontsize = 15)
matplotlib.pyplot.grid(True, which="both")
plt.xlim(0.9*min(W_2),1.1*max(W_2))
plt.ylim(1e21, 6e21)
plt.locator_params(axis='y', numticks=8)
plt.gca().xaxis.set_major_formatter(xfmt)

# EFTIR 423 K BKUN

lidur3 = pd.read_csv('lidur3',sep='\t', skiprows=5)
V_3 = lidur3[list(lidur3)[0]]
C_3 = lidur3[list(lidur3)[2]]

dCdV_3 = np.divide(C_3**(-2),V_3)
dCdV_3 = dCdV_3[1:]
W_3 = K_s*eps_0*A*C_3**(-1)
W_3 = W_3[1:] 
N_A_3 = 2/(q*K_s*eps_0*A**(2)*np.abs(dCdV_3))

plt.subplot(313)
plt.scatter(W_3, N_A_3,s = 3)
plt.yscale('log')
plt.title('Eftir 423 K bkun', fontsize = 20)
plt.xlabel('Dpt ' + r'$[\mu m$]', fontsize = 15)
plt.ylabel('btarttleiki, ' + r'$N_{A} \ [m^{-3}]$', fontsize = 15)
matplotlib.pyplot.grid(True, which="both")
plt.xlim(0.9*min(W_3),1.1*max(W_3))
plt.ylim(1e21, 6e21)
plt.locator_params(axis='y', numticks=8)
plt.gca().xaxis.set_major_formatter(xfmt)

plt.subplots_adjust(top=4, bottom=0, left=0, right=1.5, hspace=0.45,
                    wspace=0.35)



plt.savefig('MAIN_B.png', dpi=300, bbox_inches='tight')
```
