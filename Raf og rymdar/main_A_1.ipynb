```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt

# fastar til að reikna n
q = 1.602*10**(-19)
k_B = 1.38*10**(-23)
T = 298

#1N4148
fwd_1N = pd.read_csv('1nfwd.txt',sep='\t')
x_1N = fwd_1N['Voltage (V)']
y_1N = fwd_1N['Current (A)']
x_1N_to_fit = x_1N[(x_1N>0.1) & (x_1N<0.8)]
y_1N_to_fit = y_1N[(x_1N>0.1) & (x_1N<0.8)]

log_y_1N = np.log(y_1N)
log_y_1N_to_fit = np.log(y_1N_to_fit)
m_1N,c_1N = np.polyfit(x_1N_to_fit, log_y_1N_to_fit,1)
y_fit_1N = m_1N*x_1N_to_fit +c_1N
n_1N = q/(m_1N*k_B*T)
 
print('n1N', n_1N)


#BZX-55C13
fwd_BZX = pd.read_csv('bzfwd0.txt',sep='\t')
x_BZX = fwd_BZX['Voltage (V)']
y_BZX = fwd_BZX['Current (A)']

x_BZX_to_fit = x_BZX[(x_BZX>0.2) & (x_BZX<0.8)]
y_BZX_to_fit = y_BZX[(x_BZX>0.2) & (x_BZX<0.8)]

log_y_BZX = np.log(y_BZX)
log_y_BZX_to_fit = np.log(y_BZX_to_fit)
m_BZX,c_BZX = np.polyfit(x_BZX_to_fit, log_y_BZX_to_fit,1)
y_fit_BZX = m_BZX*x_BZX_to_fit +c_BZX
n_BZX = q/(m_BZX*k_B*T)
print('nBZX', n_BZX)
#BYS26-45
fwd_BYS = pd.read_csv('bysfwd.txt',sep='\t')
x_BYS = fwd_BYS['Voltage (V)']
y_BYS = fwd_BYS['Current (A)']

x_BYS_to_fit = x_BYS[(x_BYS>0.05) & (x_BYS<0.27)]
y_BYS_to_fit = y_BYS[(x_BYS>0.05) & (x_BYS<0.27)]

log_y_BYS = np.log(y_BYS)
log_y_BYS_to_fit = np.log(y_BYS_to_fit)
m_BYS,c_BYS = np.polyfit(x_BYS_to_fit, log_y_BYS_to_fit,1)
y_fit_BYS = m_BYS*x_BYS_to_fit +c_BYS
n_BYS = q/(m_BYS*k_B*T)
print('nBYS', n_BYS)
#Blár tvistur
fwd_blue = pd.read_csv('blafwd.txt',sep='\t')
x_BLUE = fwd_blue['Voltage (V)']
y_BLUE = fwd_blue['Current (A)']

log_y_BLUE = np.log(y_BLUE)
x_BLUE_to_fit_1 = x_BLUE[(x_BLUE>1.7) & (x_BLUE<2.25)]
y_BLUE_to_fit_1 = y_BLUE[(x_BLUE>1.7) & (x_BLUE<2.25)]
x_BLUE_to_fit_2 = x_BLUE[(x_BLUE>2.25) & (x_BLUE<2.6)]
y_BLUE_to_fit_2 = y_BLUE[(x_BLUE>2.25) & (x_BLUE<2.6)]

log_y_BLUE_to_fit_1 = np.log(y_BLUE_to_fit_1)
m_BLUE_1,c_BLUE_1 = np.polyfit(x_BLUE_to_fit_1, log_y_BLUE_to_fit_1,1)
y_fit_BLUE_1 = m_BLUE_1*x_BLUE_to_fit_1 +c_BLUE_1
n_BLUE_1 = q/(m_BLUE_1*k_B*T)
print('nBlue1 ', n_BLUE_1)

log_y_BLUE_to_fit_2 = np.log(y_BLUE_to_fit_2)
m_BLUE_2,c_BLUE_2 = np.polyfit(x_BLUE_to_fit_2, log_y_BLUE_to_fit_2,1)
y_fit_BLUE_2 = m_BLUE_2*x_BLUE_to_fit_2 +c_BLUE_2
n_BLUE_2 = q/(m_BLUE_2*k_B*T)
print('nBLUE2 ', n_BLUE_2)
# plot with various axes scales
plt.figure(1)

# 1N4148
plt.subplot(221)
plt.scatter(x_1N, log_y_1N,s=3)
plt.plot(x_1N_to_fit, y_fit_1N, color=[0.8500, 0.3250, 0.0980])
plt.text(x=0.2,y=-17.5,s='y ='+str(round(m_1N,4))+'x'+str(round(c_1N,4)))
#plt.yscale('log')
plt.title('p-n tvistur')
plt.grid(True)
plt.xlabel('Spenna (V)')
plt.ylabel('Logrinn af Straum')
plt.locator_params(axis='y', numticks=10)



# BZX-55C13 há íbót þröngt berasnautt bil
plt.subplot(222)
plt.scatter(x_BZX, log_y_BZX, s=3)
plt.plot(x_BZX_to_fit, y_fit_BZX, color=[0.8500, 0.3250, 0.0980])
plt.text(x=0.4,y=-17.5,s='y ='+str(round(m_BZX,4))+'x'+str(round(c_BZX,4)))
#plt.yscale('log')
plt.title('Zener tvistur')
plt.grid(True)
plt.xlabel('Spenna (V)')
plt.ylabel('Logrinn af Straum')
plt.locator_params(axis='y', numticks=10)

# BYS26-45
plt.subplot(223)
plt.scatter(x_BYS,log_y_BYS,s=3)
plt.plot(x_BYS_to_fit, y_fit_BYS, color=[0.8500, 0.3250, 0.0980])
plt.text(x=0.2,y=-8,s='y ='+str(round(m_BYS,4))+'x'+str(round(c_BYS,4)))
#plt.yscale('log')
plt.title('Schottky tvistur')
plt.grid(True)
plt.xlabel('Spenna (V)')
plt.ylabel('Logrinn af straum')
plt.locator_params(axis='y', numticks=8)

# Blár tvistur
plt.subplot(224)
plt.scatter(x_BLUE, log_y_BLUE,s = 3)
plt.plot(x_BLUE_to_fit_1, y_fit_BLUE_1,color=[0.8500, 0.3250, 0.0980])
plt.text(x=2,y=-20,s='y ='+str(round(m_BLUE_1,4))+'x'+str(round(c_BLUE_1,4)))
plt.plot(x_BLUE_to_fit_2, y_fit_BLUE_2,color='green')
plt.text(x=2.5,y=-13.5,s='y ='+str(round(m_BLUE_2,4))+'x'+str(round(c_BLUE_2,4)))
#plt.yscale('log')
plt.title('Blár ljóstvistur')
plt.grid(True)
plt.xlabel('Spenna (V)')
plt.ylabel('Logrinn af straum')
plt.locator_params(axis='y', numticks=6)
# Format the minor tick labels of the y-axis into empty strings with
# `NullFormatter`, to avoid cumbering the axis with too many labels.
#plt.gca().yaxis.set_minor_formatter(NullFormatter())
# Adjust the subplot layout, because the logit one may take more space
# than usual, due to y-tick labels like "1 - 10^{-3}"
plt.subplots_adjust(top=2, bottom=0, left=0, right=2, hspace=0.25,
                    wspace=0.35)

#plt.show() raðviðnám 
plt.savefig('MAIN_A_fwd.png', dpi=300, bbox_inches='tight')
```
